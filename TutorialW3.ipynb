```python
import numpy as np

array_2x2 = np.array([[1, 2],
                      [3, 4]])
print("2x2 Array:\n", array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print("\n Random 3x4 Array:\n", random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))
print("\n Array of Ones:\n", ones_array)
print("Array of Zeroes:\n", zeros_array)
```

```python
random_6x3 = np.random.rand(6, 3)
transpose_array = random_6x3.T
print("\n Random 6x3 Array:\n", random_6x3)
print("Transposed Array (3x6):\n", transpose_array)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)

print("\n Matrix A (n x p):\n", A)
print("Matrix B (p x m):\n", B)
print("Dot Product (n x m):\n", dot_product)
```
